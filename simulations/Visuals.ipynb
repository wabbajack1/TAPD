```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
# JSON file
with open('acc_ac_3000.json', "r") as f:
    acc_ac = json.loads(f.read())
    
with open('acc_kb_3000.json', "r") as f:
    acc_kb = json.loads(f.read())
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5), sharey=True)
fig.suptitle('Accuracy of P&C architecture')


ax1=plt.subplot(1, 3, 1)
plt.plot(acc_ac["0"], '-r', label="sgd+activecolumn")
plt.plot(acc_kb["0"], '-g', label="ewc+kb")
plt.xticks(np.arange(0, 100, step=20), rotation=20)
for i in range(0, 100+1, 20):
    plt.axvline(x=i, color="black", linestyle=":")
plt.legend()
plt.ylim(0.7, 1)

ax2=plt.subplot(1, 3, 2)
for i, y in acc_ac.items():
    plt.plot(y)
    for i in range(0, 100+1, 20):
        plt.axvline(x=i, color="black", linestyle=":")
    plt.yticks(rotation=20)
    plt.xticks(np.arange(0, 100, step=20), rotation=20)

ax3=plt.subplot(1, 3, 3)
for i, y in acc_kb.items():
    plt.plot(y)
    for i in range(0, 100+1, 20):
        plt.axvline(x=i, color="black", linestyle=":")
    plt.yticks(rotation=20)
    plt.xticks(np.arange(0, 100, step=20), rotation=20)
```
